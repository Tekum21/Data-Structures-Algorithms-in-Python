# Visualization with Matplotlib
## Matplotlib
- A Python package for multi-platform data visualization library
  - It works on many operating systems and graphics backends
- Allows easy creation of various plots including line, scattered, bar, box, etc. 
- Ease of use
  - Integration with Pandas and NumPy
  - Contains ready made default style sheets
- Extendable
  - Matplotlib is the basis for other third-party packages which extend its capabilities. These include:
    - Seaborn
    - HoloViews
    - Basemap/GeoPandas/Canopy

### Matplotlib Backends
>matplotlib targets many different use cases and output formats. Some people use matplotlib interactively from the python shell and have plotting windows pop up when they type commands. Some people embed matplotlib into graphical user interfaces like wxpython or pygtk to build rich applications. Others use matplotlib in batch scripts to generate postscript images from some numerical simulations, and still others in web application servers to dynamically serve up graphs.

>To support all of these use cases, matplotlib can target different outputs, and each of these capabilities is called a backend; the “frontend” is the user facing code, i.e., the plotting code, whereas the “backend” does all the hard work behind-the-scenes to make the figure. There are two types of backends: user interface backends (for use in pygtk, wxpython, tkinter, qt4, or macosx; also referred to as “interactive backends”) and hardcopy backends to make image files (PNG, SVG, PDF, PS; also referred to as “non-interactive backends”).

- https://matplotlib.org/faq/usage_faq.html#what-is-a-backend
- https://matplotlib.org/api/index_backend_api.html

### Two Interfaces
#### MATLAB-style Interface
- This interface is *stateful*.
  - It keeps track of the current figure and axes.
  - `plt` commands are applied on these objects.
  - Use `plt.gcf()` (get current figure) and `plt.gca()` (get current axes) methods to get references.
  - Check [here](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.html#module-matplotlib.pyplot) for more information. 

```python
plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

> Matplotlib is the whole package; matplotlib.pyplot is a module in matplotlib; and pylab is a module that gets installed alongside matplotlib.

> Pyplot provides the state-machine interface to the underlying object-oriented plotting library. The state-machine implicitly and automatically creates figures and axes to achieve the desired plot.

> **The first call to plt.plot will automatically create the necessary figure and axes to achieve the desired plot. Subsequent calls to plt.plot re-use the current axes and each add another line. Setting the title, legend, and axis labels also automatically use the current axes and set the title, create the legend, and label the axis respectively.**

> pylab is a convenience module that bulk imports matplotlib.pyplot (for plotting) and numpy (for mathematics and working with arrays) in a single name space. Although many examples use pylab, it is no longer recommended.

From: https://matplotlib.org/faq/usage_faq.html

#### Object-oriented interface
- Available for more complicated situations.
- Provides more control over the figures.
- No notion of "active" figure or axes
- Plotting functions are *methods* of explicit `Figure` and `Axes` objects
- Refer [here](https://matplotlib.org/3.2.1/api/index.html#the-object-oriented-api) for more information. 
- [matplotlib.figure.Figure](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.figure.Figure.html#matplotlib-figure-figure) is the top level container for all the plot elements. 
  - A single container for all the objects including axes, graphics, text, and labels.
- [matplotlib.axes.Axes](https://matplotlib.org/3.2.1/api/axes_api.html#matplotlib.axes.Axes) contains the figure elements and sets the coordinate system.


```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x));
```

### Adjusting the Plot: Line Colors and Styles
The `plt.plot()` function takes additional arguments used to specify how to conrol the line colors and styles.

#### Adjusting the color
Use the `color` keyword with a string value to specify a color.

If no color is specified, Matplotlib cycles through a default set of colors. 

```python
plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='g')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

#### Adjusting the line style
Use the `linestyle` keyword to adjust the line style

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

### Anatomy of Matplotlib
![](https://matplotlib.org/_images/anatomy1.png)

- Everything in Matplotlib is organized a hierarchy.
  - At the top of the hierarchy is the matplotlib “state-machine environment”
    - Provided by the matplotlib.pyplot module. 
    - At this level, simple functions are used to add plot elements (lines, images, text, etc.) to the current axes in the current figure.
  - The next level down in the hierarchy is the first level of the object-oriented interface
    - pyplot is used only for a few functions such as figure creation, and the user explicitly creates and keeps track of the figure and axes objects. 
    - pyplot ise used to create figures, and through those figures, one or more axes objects can be created. These axes objects are then used for most plotting actions.

From: https://matplotlib.org/faq/usage_faq.html
  
#### Figure and Axes
>The Figure is the top-level container in this hierarchy. It is the overall window/page that everything is drawn on. You can have multiple independent figures and Figures can contain multiple Axes.

>Most plotting ocurs on an Axes. The axes is effectively the area that we plot data on and any ticks/labels/etc associated with it. Usually we'll set up an Axes with a call to subplot (which places Axes on a regular grid), so in most cases, Axes and Subplot are synonymous.

>Each Axes has an XAxis and a YAxis. These contain the ticks, tick locations, labels, etc. In this tutorial, we'll mostly control ticks, tick labels, and data limits through other mechanisms, so we won't touch the individual Axis part of things all that much. However, it is worth mentioning here to explain where the term Axes comes from.

For more info check out https://nbviewer.jupyter.org/github/matplotlib/AnatomyOfMatplotlib/blob/master/AnatomyOfMatplotlib-Part1-Figures_Subplots_and_layouts.ipynb

## [Scatter Plots](https://en.wikipedia.org/wiki/Scatter_plot)
- Within scatter plots, data points are represented individually with a dot, circle, or other shape. 
- it is a mathematical digram using Catersian cooridnates to display values typically two variables for a set of data. 
- Used to identify correlations between variables. 
- A line of best fit can be drawn to study the relationship between the variables. 

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)
```

```python
plt.plot(x, y, 'o', color='black');
```

The third argument (in this case 'o') is a chracter representing the type of symbol used for the plotting. 

See [here](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.markers.MarkerStyle.html#matplotlib.markers.MarkerStyle) for available marker symbols. 

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

Using chracter codes along the line and color codes to plot points. 
#### Format string
A format string consists of color, maker and line:
> fmt = '[marker][line][color]'

The following example has a format string with the value '-ok'. 
- '-', solid line style (marker)
- 'o', circle marker (line)
- 'k', black (color)

Refer to the bottom of [this page](https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.plot.html) for more information. 

```python
plt.plot(x, y, '-ok')
```

```python
plt.plot(x, y, '-p', color='gray', #'p => pentagon marker'
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

### Scatter Plots with plt.scatter
Difference between `plt.scatter` from `plt.plot`
- `plt.scatter` is used to create scatter plots where the properties of each individual points can be individually controlled or mapped to data. 

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

The third argument 'c' can have a possible value from:
1. A single color format string.
2. A sequence of colors of length n.
3. A scalar or sequence of n numbers to be mapped to colors using cmap and norm.
4. A 2-D array in which the rows are RGB or RGBA

Within the above example, it is using the option #3. It is automatically mapped to a color scale drawn by `plt.colorbar()`. The 'size' argument is given in pixels.

Refer to ["Choosing Colomaps in Matplotlib"](https://matplotlib.org/tutorials/colors/colormaps.html) documentation. 

## Visualizing Errors
>For any scientific measurement, accurate accounting for errors is nearly as important, if not more important, than accurate reporting of the number itself.

Allowing a plot to convey errors and possible range of uncertainties make presented information much more complete. 

### Basic Errorbars

```python
plt.style.use('seaborn-whitegrid')
```

```python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k');
```

- `yerr` argument specifies symmetric +/- values for all data points. 
- `fmt` argument, '.' represents point marker, 'k' for color black.

## Density and Contour Plots
- Ability to display three-dimensional data in two dimensions using contours or color-coded regions. 
- Three Matplotlib functions
  - `plt.contour`, contour plots
  - `plt.contourf` for filled contour plots, filled contour fplots
  - `plt.imshow`, showing images
  

### Visualizing a Three-Dimensional Function
- Contour plot using a function z = f(x, y)
- A contour plot can be created with the `plt.contour` function. It takes three arguments:
  - a grid of *x* values, represents x position
  - a grid of *y* values, represents y position
  - a grid of *z* values, represents the contour levels
  
`np.meshgrid` builds two-dimensional grids from one-dimensional arrays.

```python
def f(x,y):
    return np.sin(x) ** 10 + np.cos(10+Y*X) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)

plt.contour(X, Y, Z, colors='black');
```

For a single color:
- negative values are represented by dashed lines. 
- positive values are represented by solid lines. 

For NumPy meshgrid, refer [here](https://docs.scipy.org/doc/numpy/reference/generated/numpy.meshgrid.html).

```python
plt.contour(X, Y, Z, 20, cmap='RdGy')
```

'RdGy' stands for Red-Gray colormap. 

Using a filled contour plot using `plt.contourf()` function. 

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

According to the colorbar, the black regions are "peaks" and the red regions are the "valleys."

Using `plt.imshow()` function to make the plot more continuous. 

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
```

[plot.clabel()](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.clabel.html) is used to add labels to line contours. 


```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)
plt.colorbar();
```

## Histograms, Binnings, and Desity
Histogram is a great tool to identify the data distribution. 

It shows the following: 
- center of the data
- spread of the data
- skewness of the data
- presence of outliers
- presence of multiple modes in the data

Reference: https://www.itl.nist.gov/div898/handbook/eda/section3/histogra.htm

```python
data = np.random.randn(1000)
```

```python
plt.hist(data)
```

```python
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

kwargs = dict(histtype='stepfilled', alpha=0.3, bins=40)

plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

`np.histogram()` function is used to count the number of points in a given bin, and not display it. 

```python
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
```
